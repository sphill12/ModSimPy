```python
import random
import matplotlib.pyplot as plt

```

## How to play street craps:

shooter rolls both dice. if they recieve a 7 or 11, they win. If they roll a 2,3 or 12 they lose. If they roll a 4,5,6,8,9,10 then this number becomes the "point number" and the shooter must now roll this number before rolling a 7 to win. If they roll a 7 before this number then they lose the pot

```python
## Define a function to get the value for 2 dice being rolled

def dice():
    return random.randint(1,6) + random.randint(1,6)

num_trials = 1000
results = {"Player 1 Round 1 Win": 0, "Player 1 Round 1 Loss": 0, "Player 1 Round 2 Win":0, "Player 1 Round 2 Loss": 0}

## Define a function for 
def craps_round():
    come_out_roll = dice()

    ## Round 1 win condition
    if come_out_roll in [7, 11]:
        results["Player 1 Round 1 Win"] += 1
        return
    ## Round 1 lose condition
    elif come_out_roll in [2,3,12]:
        results["Player 1 Round 1 Loss"] += 1
        return
    ## Move to second round if neither are satisfied
    else:
        ## Check to see if the first roll is a win
        round_2_roll = dice()
        if round_2_roll == come_out_roll:
            results["Player 1 Round 2 Win"] += 1
            return
        ## If it isn't a win, create a loop where the playereither loses (rolls 7) or wins (rolls the point number)
        while round_2_roll != come_out_roll:
            if round_2_roll == 7:
                results["Player 1 Round 2 Loss"] += 1
                return
            round_2_roll = dice()
        results["Player 1 Round 2 Win"] += 1
        return
    
```

```python
for i in range(num_trials):
    craps_round()
results
```

```python
plt.bar(results.keys(), results.values(), tick_label = list(results.keys()))
plt.xlabel("Player 1 Outcome")
plt.xticks(rotation = 45)
plt.ylabel("Frequency")
plt.title("Distribution of Craps outcomes for Player 1 Rolling 1000 times")
```
